```python
import pandas as pd
import random

file_path = '/Users/imrohu/Downloads/CESAR1-Emissions-Data-TCM-for-SBIR.xlsx'
df = pd.read_excel(file_path, sheet_name='Data Extract Offline', skiprows=7)

selected_columns = [3, 13, 16, 17, 15, 24, 35, 30]

def select_random_columns(num_columns):
    columns_to_select = [col for col in selected_columns if col != 30]
    random_columns = random.sample(columns_to_select, k=min(num_columns - 1, len(columns_to_select)))
    random_columns.append(30)
    return random_columns

num_random_columns = int(input("Enter the number: "))

random_columns = select_random_columns(num_random_columns)
print("Columns including output cloumn at last:", random_columns)

selected_values = df.iloc[:, random_columns]
print(selected_values)
```

```python
#Timesteps

import pandas as pd
file_path = '/Users/imrohu/Downloads/CESAR1-Emissions-Data-TCM-for-SBIR.xlsx'
df = pd.read_excel(file_path, sheet_name='Data Extract Offline', skiprows=7)


selected_columns = df.iloc[:, random_columns]

window_size = 10

# Create time series data
time_series_data = []
for i in range(len(selected_columns) - window_size + 1):
    time_step = selected_columns.iloc[i:i+window_size].values.flatten() 
    time_series_data.append(time_step)

time_series_df = pd.DataFrame(time_series_data, columns=[f'timestep_{i}' for i in range(window_size * selected_columns.shape[1])])

print(time_series_df)
```

```python
#Minmax scaling

from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
scaled_data = scaler.fit_transform(selected_columns)  
scaled_time_series_df = pd.DataFrame(scaled_data) 

print(scaled_time_series_df)
print(scaled_time_series_df.shape)
```

```python
#Temporalization

import numpy as np

def temporalize(X, y, lookback):
    output_X = []
    output_y = []
    for i in range(len(X)-lookback):
        t = X[i:(i+lookback), :]
        output_X.append(t)
        output_y.append(y[i+lookback])
    return output_X, output_y

timesteps = 10
X, y = temporalize(X=scaled_time_series_df.values, y=scaled_time_series_df.iloc[:, 1].values, lookback=timesteps)




n_features = scaled_time_series_df.shape[1]
X = np.array(X)

X = X.reshape(len(X), timesteps, n_features)
np.set_printoptions(suppress=True)

print("x=",X)
print(X.shape)
print("y=",y)

y = np.array(y)
print("y Shape:", y.shape)
```

```python
#Training and testing
test_split=round(len(scaled_time_series_df)*0.20)
trainX=X[:-403]
trainY=y[:-403]
testX=X[-403:]
testY=y[-403:]
```

```python
from keras.layers import Layer, Input, LSTM, Dense, RepeatVector, TimeDistributed
from keras.models import Model
import keras.backend as K
import numpy as np
from keras.layers import Layer

def build(self,input_shape):
        self.W=self.add_weight(name="att_weight",shape=(input_shape[-1],1),initializer="normal")
        self.b=self.add_weight(name="att_bias",shape=(input_shape[1],1),initializer="zeros")        
        super(attention, self).build(input_shape)
        
def call(self,x):
        et=K.squeeze(K.tanh(K.dot(x,self.W)+self.b),axis=-1)
        at=K.softmax(et)
        at=K.expand_dims(at,axis=-1)
        output=x*at
        return K.sum(output,axis=1)

class Attention(Layer):
    def __init__(self, **kwargs):
        super(Attention, self).__init__(**kwargs)

    def build(self, input_shape):
        self.W = self.add_weight(name="att_weight", shape=(input_shape[-1], 1), initializer="normal")
        self.b = self.add_weight(name="att_bias", shape=(input_shape[1], 1), initializer="zeros")
        super(Attention, self).build(input_shape)

    def call(self, x):
        et = K.squeeze(K.tanh(K.dot(x, self.W) + self.b), axis=-1)
        at = K.softmax(et)
        at = K.expand_dims(at, axis=-1)
        output = x * at
        return K.sum(output, axis=1)

    def compute_output_shape(self, input_shape):
        return input_shape[0], input_shape[-1]
    
    def get_config(self):
        return super(attention,self).get_config()


X = X.reshape(len(X), timesteps, n_features)
y = np.array(y).reshape(-1, 1)

visible = Input(shape=(timesteps, n_features))
encoder = LSTM(32, activation='relu', return_sequences=True)(visible)
encoder = LSTM(16, activation='relu', return_sequences=True)(encoder)

att_in = LSTM(units=16, return_sequences=True)(encoder)  
att_out = Attention()(att_in)
outputs=Dense(1,activation='sigmoid',trainable=True)(att_out)

decoder1 = RepeatVector(timesteps)(att_out)
decoder1 = LSTM(16, activation='relu', return_sequences=True)(decoder1)
decoder1 = LSTM(32, activation='relu', return_sequences=True)(decoder1)
decoder1 = TimeDistributed(Dense(100, activation='relu'))(decoder1)
decoder1 = TimeDistributed(Dense(n_features))(decoder1)

decoder2 = RepeatVector(1)(att_out)
decoder2 = LSTM(16, activation='relu', return_sequences=True)(decoder2)
decoder2 = LSTM(32, activation='relu', return_sequences=True)(decoder2)
decoder2 = TimeDistributed(Dense(1))(decoder2)

model = Model(inputs=visible, outputs=[decoder1, decoder2])
model.compile(optimizer='adam', loss='mse')

model.fit(X, [X, y], epochs=30, verbose=0)

y_predict = model.predict(testX, verbose=0)
print(y_predict)
```
